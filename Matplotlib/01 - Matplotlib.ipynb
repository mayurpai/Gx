```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(15)
y = x**2
z = x**3
```

```python
x.shape
```

```python
y.shape
```

```python
plt.xlabel('Number')
plt.ylabel('Square Of Number')
plt.title('Number Vs. Square Of Number')
plt.plot(x, y)
plt.show()
```

```python
plt.xlabel('Number', fontdict={'size': 15, 'color':'red', 'family':'serif'})
plt.ylabel('Square/Cube Of Number', fontdict={'size': 15, 'color':'green', 'family':'arial'})
plt.title('Number Vs. Square/Cube Of Number', fontdict={'size': 15, 'color':'blue'})
plt.plot(x, y,  marker='o', linestyle='dashed', linewidth=2, ms="5",
         mec='yellow', mfc='black', label='line', alpha=0.5, color='red')
plt.plot(x, z,  marker='*', linestyle='solid', linewidth=2, ms="5",
         mec='orange', mfc='black', label='line', alpha=0.5, color='green')
plt.grid(axis='y', linestyle='dashdot', linewidth=0.5, color='green')
plt.grid(axis='x', linestyle='-', linewidth=0.5, color='red')
plt.legend(['Squares', 'Cubes'], loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.subplot(2,2,1)
plt.plot(x, y, label='Squared Numbers')
plt.legend(['Squares'], edgecolor='black', facecolor='white', framealpha=1, fancybox=True, shadow=True, fontsize=10)
plt.xlabel('Numbers')
plt.ylabel('Squared Numbers')
plt.title('Numbers Vs. Square Numbers')
plt.show()

plt.subplot(2,2,2)
plt.plot(x, z, label='Cubed Numbers')
plt.legend(['Cubes'])
plt.xlabel('Numbers')
plt.ylabel('Cubes Numbers')
plt.title('Numbers Vs. Cubes Numbers')
plt.show()

plt.subplot(2,2,3)
plt.plot(y, z, label='Squared Numbers')
plt.ylabel('Numbers')
plt.xlabel('Squared Numbers')
plt.title('Square Numbers Vs. Numbers')
plt.show()

plt.subplot(2,2,4)
plt.plot(z, x, label='Cubed Numbers')
plt.ylabel('Numbers')
plt.xlabel('Cubes Numbers')
plt.title('Cubes Numbers Vs. Numbers')
plt.show()
```

```python
import random
x_scatter = np.arange(10)
y_scatter = np.arange(0, 100, 10)
colors = ['red', 'blue', 'green', 'yellow', 'black', 'orange', 'brown', 'purple', 'cyan', 'pink']
size = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
random.shuffle(x_scatter)
plt.scatter(x_scatter, y_scatter,c=colors, marker='o', linewidths=1, s=size, alpha=0.5, cmap='rdbu')
plt.colorbar()
random.shuffle(x_scatter)
plt.scatter(x_scatter, y_scatter,c=colors, marker='*', linewidths=10, s=size, cmap='RdYlGn')
plt.colorbar()
plt.show()
```

```python
x = ['a', 'b', 'c']
y = [1, 2, 3]
plt.subplot(1, 2, 1)
plt.bar(x, y, color = colors, width=.5);
plt.subplot(1, 2, 2)
plt.barh(x, y, color = colors, height=.5);
plt.show()
```

```python
years = ['2017', '2018', '2019']
placemenets = [100, 110, 120]
cse = [10, 20, 30]
ise = [40, 50, 60]
ece = [20, 80, 90]

plt.bar(years, ece, color='g')
plt.bar(years, ise, color='r')
plt.bar(years, cse, color='b')
plt.legend(['ECE', 'ISE', 'CSE'])
plt.title('Placemenets Comparision')
plt.show()
```

```python
plt.title('Performance')
plt.pie(placemenets,  labels=years, startangle=90, explode=[0.2, 0, 0], shadow=True, colors=colors, autopct="%2.3f%%")
plt.figure(figsize=(8,12))
plt.show()
```
