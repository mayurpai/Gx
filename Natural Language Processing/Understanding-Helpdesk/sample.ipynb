```python
import pandas as pd
import numpy as np
import pyodbc
import email
import html2text
import re
```

```python
# Read the CSV file into a DataFrame
csv_file_path = './Sep 2020-2022.csv'
dataframe = pd.read_csv(csv_file_path)
print(dataframe.head())
```

```python
# Connect to the database (sql server)
db_connection = pyodbc.connect('Driver={SQL Server};'
                               'Server=GXWSBCLAIMDB01;'
                               'Database=ManageEngine;'
                               'Trusted_Connection=yes;')

cursor = db_connection.cursor()
```

```python
# Fetch description data for workorder IDs using DataFrame Request ID
workorder_ids = dataframe['Request ID'].tolist()
workorder_ids = tuple(workorder_ids)
print(workorder_ids)
```

```python
query = f"SELECT WORKORDERID,TITLE,DESCRIPTION FROM WorkOrder WHERE WORKORDERID IN {workorder_ids}"
cursor.execute(query)
data = cursor.fetchall()
```

```python
# Create a new DataFrame using the query results
result_dataframe = pd.DataFrame(
    np.array(data), columns=['WORKORDERID', 'TITLE', 'DESCRIPTION'])
print(range(len(result_dataframe)))


result_dataframe["DESCRIPTION"] = result_dataframe["DESCRIPTION"].replace(
    to_replace=["&nbsp;", "&amp;", "&lt;", "&gt;", "&quot;", "&n...", "&nb...", "&#39;", "&#39", "&#...", "&..."], value='', regex=True)
print(result_dataframe)
result_dataframe.to_csv('workorder_data.csv', index=False)

# # for x in data:
# #     print(x[0])
# #     print('\n')
# data1 = [x[0] for x in data]
# # print(data1,end='\n')
# data2 = []
# for description in data1:
#     message = email.message_from_string(description)
#     content = message.get_payload()
#     content = html2text.html2text(content)
#     content = re.sub("(From:(.*?))", "", content).strip()
#     data2.append(content)
# dataframe['Description'] = data2
# data = pd.read_sql(query, db_connection)
# print(data)
```
